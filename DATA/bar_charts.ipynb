```python
# Dependencies and Setup
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
```

```python
internal_candidates_pd = pd.read_csv("internal_candidates.csv")
internal_candidates_pd
```

```python
#objects = internal_candidates_pd.groupby[{"birth_date"}].sum()
datedf = pd.to_datetime(internal_candidates_pd["birth_date"]).value_counts().sort_index().reset_index()
datedf.reset_index(inplace=True)
datedf = datedf.rename(columns = {'index':'Birthdate'})
datedf
```

```python
plt.plot(datedf['Birthdate'],datedf['birth_date'])
plt.rcParams["figure.figsize"] = [15,3]
plt.xlabel("Birthdate")
plt.ylabel("Employees")
plt.title('Employee Birthday Chart')
plt.show()
```

```python
all_employees_pd = pd.read_csv("employees.csv")
all_employees_pd
```

```python
#objects = internal_candidates_pd.groupby[{"birth_date"}].sum()
datedf = pd.to_datetime(all_employees_pd["birth_date"]).value_counts().sort_index().reset_index()
datedf.reset_index(inplace=True)
datedf = datedf.rename(columns = {'index':'Birthdate'})
datedf
```

```python
plt.plot(datedf['Birthdate'],datedf['birth_date'])
plt.rcParams["figure.figsize"] = [15,3]
plt.xlabel("Birthdate")
plt.ylabel("Employees")
plt.title('All Birthdays Birthday Chart')
plt.show()
```
