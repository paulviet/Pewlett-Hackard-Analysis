```python
# Dependencies and Setup
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
internal_candidates_pd = pd.read_csv("internal_candidates.csv")
internal_candidates_pd
```

```python
#objects = internal_candidates_pd.groupby[{"birth_date"}].sum()
datedf = pd.to_datetime(internal_candidates_pd["birth_date"]).value_counts().sort_index().reset_index()
datedf
```

```python
plt.plot(datedf['index'],datedf.index.values)
plt.rcParams["figure.figsize"] = [10,3]
plt.xlabel("Birthdate")
plt.ylabel("Employees")
plt.show()
```
